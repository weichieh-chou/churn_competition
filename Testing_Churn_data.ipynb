**Issue with churn:**\
Because we have to predict if users are likely to churn in the next 10 days we can't simply check if a user churned in total. Instead we have to chosse a cut off data with at least 10 days after the date and see if users churned in those 10 days. But that would mean that we will loose a lot of data, since users who churned before that won't be taken into account.\
**Solution to start with:** We will first start with a simple single cutoff on november 10. Then we'll do everything(train our model) and establish a baseline. After that we will try a Multi Window approach where we will cut off on multiple days and then stack these dataframes on top of each other. (users could be in the dataset multiple times but that could be an advantage because we can see changes in their pattern which is especially useful when they churn later)
