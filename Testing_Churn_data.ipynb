```python
print("Unique Value Count")
few_unique_columns = []

for column_name in df_churn.columns:
    unique_count = df_churn[column_name].nunique()
    print(f"{column_name}: {unique_count}")
    if unique_count < 20:
        few_unique_columns.append(column_name)
```

NaN and others:\
length can be NaN, song and artist can be None

```python
for column_name in few_unique_columns:
    unique_values = df_churn[column_name].unique()
    print(f"{column_name}: {unique_values}")
```

```python
max_time = df_churn["time"].max()
min_time = df_churn["time"].min()
print(f"Time min and max = {min_time} - {max_time}")
```

```python
max_registration = df_churn["registration"].max()
min_registration = df_churn["registration"].min()
print(f"Registration min and max = {min_registration} - {max_registration}")
```
